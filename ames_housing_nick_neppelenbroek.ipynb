```python
#step 1 import packages and classes
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
f, ax = plt.subplots(figsize = (15,10))
sns.regplot(x=df['Gr Liv Area'], y=np.log(df['SalePrice']), ax = ax)
```

```python
#step 2 provide data x-axis and reshape in 1 column
X = np.array(df['Gr Liv Area']).reshape((-1,1))
print(X)
```

```python
#step 2 provide data y-axis
y = np.log(df['SalePrice'])
print(y)
```

```python
#step 3 create a model
model = LinearRegression().fit(X, y)
```

```python
#Step 5: Get results;calculate square of residuals, intercept and slope
r_sq = model.score(X,y)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
#step 5: predict x from linear regression model
y_pred = model.predict(X)
print('predicted outcome:', y_pred, sep='\n')
```
